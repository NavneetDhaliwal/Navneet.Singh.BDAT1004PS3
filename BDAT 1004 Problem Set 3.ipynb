```python
# Importing library 
import pandas as panda

# Importing the dataset from provided URL using Pandas
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
UsersData = panda.read_csv(url, sep='|')

# Grouping the data by occupation and calculating the mean age for each occupation group
MeanAgeByOccupation = UsersData.groupby('occupation')['age'].mean()

# Calculating the male ratio per occupation and sorting it in descending order
# Grouping the data by occupation, calculating the proportion of males in each occupation group,
# and then sorting the results in descending order
MaleRatioByOccupation = UsersData.groupby('occupation')['gender'].apply(lambda x: (x == 'M').mean()).sort_values(ascending=False)

# Calculating the minimum and maximum ages per occupation
# Grouping the data by occupation and calculating the minimum and maximum age for each occupation group
MinAgeByOccupation = UsersData.groupby('occupation')['age'].min()
MaxAgeByOccupation = UsersData.groupby('occupation')['age'].max()

# Calculating the mean age for each combination of occupation and gender
# Grouping the data by occupation and gender, and calculating the mean age for each combination
MeanAgeByOccupationAndGender = UsersData.groupby(['occupation', 'gender'])['age'].mean()

# Calculating the percentage of women and men for each occupation
# Grouping the data by occupation and counting the total number of individuals in each occupation,
# then calculating the percentage of women and men in each occupation group
TotalIndividualsByOccupation = UsersData.groupby('occupation')['gender'].count()
PercentageWomenByOccupation = (UsersData[UsersData['gender'] == 'F'].groupby('occupation')['gender'].count() / TotalIndividualsByOccupation) * 100
PercentageMenByOccupation = (UsersData[UsersData['gender'] == 'M'].groupby('occupation')['gender'].count() / TotalIndividualsByOccupation) * 100

# Printing the results for analysis
print("Mean age by occupation:")
print(MeanAgeByOccupation)
print("\nMale ratio by occupation:")
print(MaleRatioByOccupation)
print("\nMinimum age by occupation:")
print(MinAgeByOccupation)
print("\nMaximum age by occupation:")
print(MaxAgeByOccupation)
print("\nMean age by combination of occupation and gender:")
print(MeanAgeByOccupationAndGender)
print("\nPercentage of women by occupation:")
print(PercentageWomenByOccupation)
print("\nPercentage of men by occupation:")
print(PercentageMenByOccupation)
```

```python
# Importing libraries - pandas for data manipulation and analysis
import pandas as panda
import numpy as nump

# Importing the dataset from the file 'wind.txt'
Data = panda.read_csv('wind.txt', delim_whitespace=True, na_values='NaN')

# Replacing the first 3 columns with a proper datetime index to facilitate time-based analysis
DateColumns = ['Yr', 'Mo', 'Dy']
DateFormat = "%y %m %d"
Data['Date'] = panda.to_datetime(Data[DateColumns].astype(str).agg(' '.join, axis=1), format=DateFormat)
Data = Data.set_index('Date')
Data = Data.drop(DateColumns, axis=1)

# Correcting year 2061 - a potential data error - to be compatible with datetime representation
def FixYear(year):
    if year > 2000:
        return year - 100
    else:
        return year

Data.index = Data.index.map(lambda x: x.replace(year=FixYear(x.year)))

# Converting the index to datetime format for accurate date-based analysis
Data.index = panda.to_datetime(Data.index, format='%y-%m-%d')

# Calculating the count of missing values for each location across the entire record
MissingValuesPerLocation = Data.isnull().sum()

# Computing the total count of non-missing values in the dataset
NonMissingValuesCount = Data.notnull().sum().sum()

# Calculating the average windspeed across all locations and time points
MeanWindspeed = Data.mean().mean()

# Creating a DataFrame named 'LocationData' to store statistics (min, max, mean, std) of windspeeds at each location over all days
LocationData = panda.DataFrame()
LocationData['Min'] = Data.min()
LocationData['Max'] = Data.max()
LocationData['Mean'] = Data.mean()
LocationData['Std'] = Data.std()

# Creating a DataFrame named 'DailyFigures' to store statistics (min, max, mean, std) of windspeeds across all locations for each day
DailyFigures = panda.DataFrame()
DailyFigures['Min'] = Data.min(axis=1)
DailyFigures['Max'] = Data.max(axis=1)
DailyFigures['Mean'] = Data.mean(axis=1)
DailyFigures['Std'] = Data.std(axis=1)

# Calculating the average windspeed in January for each location, considering both January 1961 and January 1962
JanuaryAverage = Data[Data.index.month == 1].mean()

# Downsampling the data to a yearly frequency for each location
YearlyData = Data.resample('Y').mean()

# Downsampling the data to a monthly frequency for each location
MonthlyData = Data.resample('M').mean()

# Downsampling the data to a weekly frequency for each location
WeeklyData = Data.resample('W').mean()

# Calculating statistics (min, max, mean, std) of windspeeds across all locations for each of the first 52 weeks (assuming the first week starts on January 2, 1961)
WeeklyFigures = panda.DataFrame()
WeeklyFigures['Min'] = WeeklyData.iloc[:52].min()
WeeklyFigures['Max'] = WeeklyData.iloc[:52].max()
WeeklyFigures['Mean'] = WeeklyData.iloc[:52].mean()
WeeklyFigures['Std'] = WeeklyData.iloc[:52].std()

# Analysis Results Display

# Displaying missing values for each location
print("Missing Values Analysis:")
print("Missing Values for Each Location:")
print(MissingValuesPerLocation)

# Displaying total count of non-missing values
print("\nNon-Missing Values Overview:")
print("Total Non-Missing Values:", NonMissingValuesCount)

# Displaying mean windspeeds across all locations and times
print("\nMean Windspeeds Overview:")
print("Mean Windspeeds over All Locations and Times:", MeanWindspeed)

# Displaying statistics of windspeeds at each location
print("\nLocation-wise Windspeed Statistics:")
print("Statistics of Windspeeds at Each Location:")
print(LocationData)

# Displaying statistics of windspeeds at each location for each day
print("\nDaily Windspeed Statistics per Location:")
print("Statistics of Windspeeds at Each Location for Each Day:")
print(DailyFigures)

# Displaying average windspeed in January for each location
print("\nJanuary Windspeed Analysis:")
print("Average Windspeed in January for Each Location:")
print(JanuaryAverage)

# Displaying yearly frequency of windspeed data for each location
print("\nYearly Frequency Analysis:")
print("Yearly Frequency for Each Location:")
print(YearlyData)

# Displaying monthly frequency of windspeed data for each location
print("\nMonthly Frequency Analysis:")
print("Monthly Frequency for Each Location:")
print(MonthlyData)

# Displaying weekly frequency of windspeed data for each location
print("\nWeekly Frequency Analysis:")
print("Weekly Frequency for Each Location:")
print(WeeklyData)

# Displaying statistics of windspeeds across all locations for each of the first 52 weeks
print("\nWeekly Windspeed Statistics for Initial 52 Weeks:")
print("Figures for Windspeeds across All Locations for the First 52 Weeks:")
print(WeeklyFigures)
```

# Question 9

Make sure that the pie chart is an even circle, labels the categories, and shows the 
percentage breakdown of the categories.

```python
# Importing the pandas library 
import pandas as panda

# Importing the 'pyplot' 
import matplotlib.pyplot as matplot

# Step 1: Loading the Roman Emperor data from the CSV file
FilePath = "C:/Users/navne/OneDrive/Desktop/roman-emperor-reigns.csv"
EmperorData = panda.read_csv(FilePath)

# Step 2: Counting the number of emperors who were assassinated
AssassinatedEmperorCount = EmperorData[EmperorData['Cause_of_Death'].str.contains('Assassinated', case=False)].shape[0]

# Step 3: Counting the number of emperors who did not die by assassination
NotAssassinatedEmperorCount = EmperorData[~EmperorData['Cause_of_Death'].str.contains('Assassinated', case=False)].shape[0]

# Step 4: Creating a pie chart to visualize the distribution of Roman Emperors by cause of death
matplot.figure(figsize=(8, 8))

# Labels for the two categories: Assassinated and Not Assassinated
CategoryLabels = ['Assassinated', 'Not Assassinated']

# Data representing the counts of the two categories
CategoryCounts = [AssassinatedEmperorCount, NotAssassinatedEmperorCount]

# Colors for the pie chart segments: Red for Assassinated and Grey for Not Assassinated
CategoryColors = ['red', 'grey']

# Creating the pie chart with percentage annotations, starting angle at 140 degrees
matplot.pie(CategoryCounts, labels=CategoryLabels, colors=CategoryColors, autopct='%1.1f%%', startangle=140)

# Adding title to the pie chart
matplot.title('Distribution of Roman Emperors by Cause of Death')

# Ensuring equal aspect ratio for a circular pie chart
matplot.axis('equal')

# Displaying the pie chart
matplot.show()
```
